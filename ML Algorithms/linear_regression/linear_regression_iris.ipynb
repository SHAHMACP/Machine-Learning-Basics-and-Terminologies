```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris
```

##### Simple linear regression with single variable

```python
iris=iris[['petal_length','petal_width']]
iris
```

y=mx+c , Let, x= petal_length and y=petal_width

```python
x=iris[['petal_length']]
y=iris[['petal_width']]
```

##### To apply linear regression , there should be a linear relation between the dependent and independent variable.

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

##### Scatter plot shows a positive correlation between petal length and petal width

##### Splitting the dataset to train and test the model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=23)
```

```python
x_train
```

```python
print(type(x_train))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#creating object for LinearRegression class
lr=LinearRegression()
```

After calling fit(), the linear regression model will be trained using the provided training data, and it will learn the relationship between the features and the target values. The model will adjust its parameters (coefficients) to minimize the difference between the predicted values and the actual target values.

```python
#  train the model with the provided training data
lr.fit(x_train,y_train)
```

```python
# y=mx+c c=intercept
c=lr.intercept_
c
```

```python
# y=mx+c m=slope
m=lr.coef_
m
```

```python
# predicted values using predict()
predicted_train=lr.predict(x_train)
```

After executing this line of code, predicted_train will contain the predicted values for the target variable based on the features in the training data. These predicted values can then be compared with the actual target values (y_train) to evaluate the performance of the model or for any further analysis.

### Evaluation

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predicted_test)
mse
```

##### Lower values of MSE indicate better model performance, as they imply that the model's predictions are closer to the actual values on average.
##### MSE value of 0.0391 suggests relatively low error

```python
rmse=np.sqrt(mse)
rmse
```

##### A Root Mean Squared Error (RMSE) value of 0.1977 means that, on average, the difference between the actual petal widths and the predicted petal widths is approximately 0.1977 units.

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted_test)
r2
```

##### Higher R-squared values indicate better fit of the model to the data. A score of 0.9307 suggests that the linear regression model fits the data very well

R-squared values range from 0 to 1

```python
# lr.score(x_train, y_train) calculates the R-squared score on the training data
# This represents how well the model fits the training data.
lr.score(x_train,y_train)
```

```python
# lr.score(x_test, y_test) calculates the R-squared score on the test data .
# This represents how well the model generalizes to unseen data.
lr.score(x_test,y_test)
```

##### The score() method returns the R-squared value by default for regression models like linear regression. This value ranges between 0 and 1, where a higher value indicates a better fit of the model to the data.

##### By comparing the R-squared scores on both training and test data, you can assess the model's performance and whether it is overfitting or underfitting. If the R-squared score is significantly higher on the training data compared to the test data, it may indicate overfitting. Conversely, if the scores are similar but low, it may indicate underfitting.

```python
#cross validation
from sklearn.model_selection import cross_val_score
scores=cross_val_score(lr,x_train,y_train,cv=5)
```
