```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

labels = ['P1','P2','P3','P4','P5']
X = np.array([[1,2],[3,4],[6,8],[7,9],[9,10]])  # dummy spatial data

Z = linkage(X, method='single')

plt.figure(figsize=(8,5))
dendrogram(Z, labels=labels)
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
# -------------------------------------------------
# Hierarchical Clustering from a Distance Matrix
# Using Complete Linkage
# -------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# -------------------------------------------------
# Step 1: Define the upper triangular distances
# Order of pairs in condensed form for 5 points:
# (P1,P2), (P1,P3), (P1,P4), (P1,P5),
# (P2,P3), (P2,P4), (P2,P5),
# (P3,P4), (P3,P5),
# (P4,P5)
# -------------------------------------------------
condensed_distances = np.array([
    9,   # P1-P2
    3,   # P1-P3
    6,   # P1-P4
    11,  # P1-P5
    7,   # P2-P3
    5,   # P2-P4
    10,  # P2-P5
    9,   # P3-P4
    2,   # P3-P5
    8    # P4-P5
])

labels = ['P1', 'P2', 'P3', 'P4', 'P5']

# -------------------------------------------------
# Step 2: Perform hierarchical clustering
# -------------------------------------------------
Z = linkage(condensed_distances, method='complete')

# -------------------
```

```python
# --------------------------------------------
# Step 3: Plot the dendrogram
# --------------------------------------------
plt.figure(figsize=(8, 5))
dendrogram(Z, labels=labels, distance_sort='ascending', color_threshold=0)
plt.title("Hierarchical Clustering Dendrogram (Complete Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance (Level)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# -------------------------------------------------
# Hierarchical Clustering from Distance Matrix
# Using Single Linkage Method
# -------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# -------------------------------------------------
# Step 1: Define the upper triangular distances
# Order of pairs in condensed form for 5 points:
# (P1,P2), (P1,P3), (P1,P4), (P1,P5),
# (P2,P3), (P2,P4), (P2,P5),
# (P3,P4), (P3,P5),
# (P4,P5)
# -------------------------------------------------
condensed_distances = np.array([
    9,   # P1-P2
    3,   # P1-P3
    6,   # P1-P4
    11,  # P1-P5
    7,   # P2-P3
    5,   # P2-P4
    10,  # P2-P5
    9,   # P3-P4
    2,   # P3-P5
    8    # P4-P5
])

labels = ['P1', 'P2', 'P3', 'P4', 'P5']

# -------------------------------------------------
# Step 2: Perform hierarchical clustering
# -------------------------------------------------
Z = linkage(condensed_distances, method='single')

# -------------------------------------------------
# Step 3: Plot the dendrogram
# -------------------------------------------------
plt.figure(figsize=(8, 5))
dendrogram(Z, labels=labels, distance_sort='ascending', color_threshold=0)
plt.title("Hierarchical Clustering Dendrogram (Single Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance (Level)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
