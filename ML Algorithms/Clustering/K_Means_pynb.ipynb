```python
import pandas as pd
df=pd.read_csv("/content/Mall_Customers - Mall_Customers - Mall_Customers - Mall_Customers.csv")
df
```

```python
x=df.iloc[:,[3,4]].values
y=df.iloc[:,3].values
```

```python
from sklearn.cluster import KMeans

inertia=[]
k_values=range(1,11)
for k in k_values:
    kmeans=KMeans(n_clusters=k,init='k-means++',random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
```

```python
# plot the elbow cu rve
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(k_values,inertia,'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans_k=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans_k.fit_predict(x)
```

```python
plt.figure(figsize=(8,6))
colors=['red','blue','green','cyan','magenta']
for i in range(5):
  plt.scatter(x[y_kmeans==i,0],x[y_kmeans==i,1],
              s=100,c=colors[i],label=f"Cluster{i+1}")

plt.scatter(kmeans_k.cluster_centers_[:,0],kmeans_k.cluster_centers_[:,1],
              s=300,c='yellow',label='Centroids')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters of Customers')
plt.legend()
plt.show()
```
