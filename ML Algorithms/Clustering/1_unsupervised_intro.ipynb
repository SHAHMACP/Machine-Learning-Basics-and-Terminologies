## UNSUPERVISED LEARNING


Unsupervised learning is a type of machine learning where the algorithm analyzes unlabeled data to find patterns or groupings. It helps identify clusters, anomalies, or hidden structures without predefined categories.

### Unsupervised learning algorithms
1. Clustering
2. Association Rule Learning
3. Dimensionality Reduction

**CLUSTERING**

Clustering in unsupervised machine learning groups unlabeled data based on similarities, identifying patterns without prior knowledge. It processes raw data into meaningful clusters, helping reveal relationships. For example, it can group clients based on input parameters without predefined output labels.

common clustering algorithmsa are as follows:
1. Hierarchical clustering
2. K-Means clustering
3. Gaussian Mixture models
4. Density-Based clustering (DBSCAN)



**DIMENSIONALITY REDUCTION**

Dimensionality reduction reduces dataset features while preserving key information, improving ML performance and visualization. For example, a 100-feature student dataset can be simplified to height and grades for easier analysis.

common dimensionality reduction algorithms are:
1. Linear Descriminant Analysis(LDA)
2. Principal component Analysis(PCA)
3. Isomap
4. Non negative Matrix Facterization(NMF)
5. Independent Component Analysis
6. Locally Linear Embedding (LLE)
7. Singular Value Decomposition (SVD)

**CHALLENGES OF UNSUPERVISED LEARNING**

1. Lack of ground truth
2. Choosing the right algorithm
3. defining the number of clusters
4. Scalability
5. High dimentional data
6. Interpreting results
7. Sensitive to noise and outliers

**APPLICATIONS OF UNSUPERVISED LEARNING**

1. Customer segmentation
2. Recommendation systems
3. Image and video analysis
4. Dimensionality reduction
5. Topic modeling
