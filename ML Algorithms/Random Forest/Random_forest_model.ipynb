```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
df = pd.read_csv('/content/updated_random_forest_classification_dataset.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = label_encoder.fit_transform(df[i])
```

```python
x = df.drop('Churn' , axis = 1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
model = RandomForestClassifier(n_estimators=100,        # Number of trees
    criterion="gini",        # Splitting criterion (entropy can also be used)
    max_depth=5,             # Maximum depth of each tree
    min_samples_split=2,     # Minimum samples required to split a node
    min_samples_leaf=1,      # Minimum samples required in a leaf node
    max_features="sqrt",     # Number of features to consider for best split
    bootstrap=True,          # Use bootstrap sampling
    random_state=42,         # Ensure reproducibility
    verbose=1 )
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
