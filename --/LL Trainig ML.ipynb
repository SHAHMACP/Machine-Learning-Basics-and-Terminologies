# Standardisation (Scaling)

To make all data in each feature to a standard range scale.


In Gender, highest value is 1 and lowest is 0. Range is 1-0=1. In Age maximum is 31 and lowest is 19. So Range is 12. These should be in a common scale. So use StandardScaler

scaled data z=(data-mean)/std
