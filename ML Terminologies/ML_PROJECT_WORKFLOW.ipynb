# **Missing value Handling**

Remove missing values

```python
print("Before missing value removal, number of rows with missing values are \n",datacopy.isnull().sum())
print("Shape of dataset before missing value removal",datacopy.shape)
datacopy = datacopy.dropna()
print("\nAfter missing value removal, number of rows with missing values are \n",datacopy.isnull().sum())
print("Shape of dataset after missing value removal",datacopy.shape)
```

Missing value imputation with SimpleImputer

```python
datacopy = data.copy()
```

```python
print("Before SimpleImputer, number of rows with missing values are \n",datacopy.isnull().sum())
print("Shape of dataset before SimpleImputer",datacopy.shape)

from sklearn.impute import SimpleImputer

categorical_imputer = SimpleImputer(strategy='most_frequent')
numerical_imputer = SimpleImputer(strategy='mean')

for column in datacopy.columns:
    if datacopy[column].dtype in ['int64', 'float64']:
        datacopy[column] = numerical_imputer.fit_transform(datacopy[[column]]).ravel()
    else:
        datacopy[column] = categorical_imputer.fit_transform(datacopy[[column]]).ravel()

print("After SimpleImputer, number of rows with missing values are \n",datacopy.isnull().sum())
print("Shape of dataset after SimpleImputer",datacopy.shape)
```

**Using VarianceThreshold   - Option 1**

ðŸ“Œ When to Use VarianceThreshold
âœ… Use Case

  - You want to remove features with very little variation across rows.

  - Works well for both numerical and categorical (encoded) data.

  - Itâ€™s an unsupervised method â€” doesnâ€™t consider the target variable y.

```python
from sklearn.feature_selection import VarianceThreshold

# Assuming x is your DataFrame
# Set a threshold â€” e.g., 0.01 for very low variance
selector = VarianceThreshold(threshold=0.5)
x_selected = selector.fit_transform(x)

# Get selected column names
selected_columns = x.columns[selector.get_support()]
print("Selected Features based on Variance Threshold:")
print(selected_columns)
```

**Univariate Selection - chi2  - Option 2**


When to Use chi2

  - Input (X): Non-negative numerical features (commonly categorical features encoded as 0/1 or ordinal integers).

  - Target (y): Categorical (e.g., binary classification).

  - Use Case: Find features that are statistically significant to the target.



```python
from sklearn.feature_selection import SelectKBest,chi2
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x,y)
feature_names = list(x.columns)
feature_scores = pd.DataFrame({'Feature': feature_names, 'Score':fit.scores_})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)
feature_scores
```

**Univariate Selection - mutual_info_classif - Option 3**

When to use

   -  You have categorical or numerical features
   -  You want to find non-linear relationships with the target
   -  You're working on classification problems

```python
from sklearn.feature_selection import mutual_info_classif
scores = mutual_info_classif(x, y)
feature_names = list(x.columns)
feature_scores = pd.DataFrame({'Feature': feature_names, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)
feature_scores
```

Feature distribution ploting and skewness checking

*  -0.5 < Symmetrical < +0.5
*  -1.0 < Moderate Negatively skewed < -0.5
*  0.5 < Moderate Positively skewed < +1.0
*  Negatively skewed < -1.0
*  Positively skewed > +1.0





```python
import numpy as np
import scipy.stats as stats

mode = "exp"

if(mode=="log"):
    datacopy[feature] = np.log(datacopy[feature])
elif(mode=="reci"):
    datacopy[feature] = 1/datacopy[feature]
elif (mode == "sqrt"):
    datacopy[feature] = np.sqrt(datacopy[feature])
elif (mode == "exp"):
    datacopy[feature] = datacopy[feature]**(1/1.2)
elif (mode == "box"):
    datacopy[feature], parameters = stats.boxcox(datacopy[feature])

#datacopy[feature] = np.log(datacopy[feature])
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X = data[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
                 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
                 'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y = data[['Loan_Status']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4000)
print(X_train.shape)
print(X_test.shape)
```

# **Model Selection**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import classification_report,confusion_matrix
from sklearn import metrics

from sklearn import model_selection

models = [
        ('LogReg', LogisticRegression()),
        ('RF', RandomForestClassifier()),
        ('KNN', KNeighborsClassifier()),
        ('SVM', SVC()),
        ('MLP', MLPClassifier())
    ]

scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
dfs     = []
target_names = ['1','2']
for name, model in models:
  kfold = model_selection.KFold(n_splits=5, shuffle=True, random_state=90210)
  cv_results = model_selection.cross_validate(model, X_train, y_train.values.ravel(), cv=kfold, scoring=scoring)
  clf = model.fit(X_train, y_train.values.ravel())
  y_pred = clf.predict(X_test)
  print("\n"+name)
  print(classification_report(y_test, y_pred, target_names=target_names))
  confusionMatrix = confusion_matrix(y_test, y_pred)
  cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusionMatrix, display_labels=['1','2'])
  cm_display.plot()
  plt.show()
  this_df = pd.DataFrame(cv_results)
  this_df['model'] = name
  dfs.append(this_df)

finalData = pd.concat(dfs, ignore_index=True)
print(finalData)
```

# **Hyper parameter tuning**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
param_grid = {
    'n_estimators': [25,100],
    'max_features': ['sqrt', 'log2'],
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_depth': [9,10],
    'max_leaf_nodes': [6,11]
}
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=cv)
grid_search.fit(X_train, y_train.values.ravel())
grid_search.best_estimator_
```

```python
print(grid_search.best_params_)
```

Plot tree

```python
from sklearn.tree import plot_tree
rf_best = grid_search.best_estimator_
plt.figure(figsize=(80, 40))
plot_tree(rf_best.estimators_[3], feature_names=X_train.columns, class_names=['1', '2'],filled=True);
plt.show()
```

# **Final Model Training  - Optional**

```python
from sklearn.ensemble import RandomForestClassifier
import pickle
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

classifier_rf = RandomForestClassifier(criterion='gini',
                                       max_depth=10,
                                       max_features='sqrt',
                                       max_leaf_nodes=11,
                                       n_estimators=100)
classifier_rf.fit(X_train, y_train)
y_pred = classifier_rf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
ax = sns.heatmap(cm, annot=True, fmt='d').set_title('Confusion Metrix')
plt.show()
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
pickle.dump(classifier_rf, open('model.pkl', 'wb'))
```
